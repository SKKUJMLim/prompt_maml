```python
import matplotlib.pyplot as plt
import numpy as np
```

# 5-way 1-shot

```python
sizes = np.array([0, 5, 10, 15, 20, 25, 30])

padding_mean = np.array([47.92, 55.52, 55.18, 55.13, 54.75, 55.51, 55.03])
fixed_mean   = np.array([47.92, 54.92, 54.87, 55.32, 54.67, 54.79, 54.37])
random_mean  = np.array([47.92, 55.08, 54.25, 55.01, 55.35, 54.88, 54.56])

#  
plt.figure(figsize=(8, 5))
plt.plot(sizes, fixed_mean, '-s', label='Fixed Patch', linewidth=2.5)
plt.plot(sizes, padding_mean, '-o', label='Padding',linewidth=2.5)
plt.plot(sizes, random_mean, '-^', label='Random Patch', linewidth=2.5)

# , , 
# plt.title('Impact of Prompt Size and Template on miniImageNet (5-way 5-shot)', fontsize=12)
plt.ylim(47, 57)
plt.xlabel('Prompt Size', fontsize=16)
plt.ylabel('Accuracy (%)', fontsize=16)
plt.xticks(sizes, fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=15)

#    
plt.tight_layout()
plt.savefig("5way_1shot_promp_design.png", dpi=200, bbox_inches='tight')
plt.show()
```
