```python
import matplotlib.pyplot as plt
import numpy as np
```

# 5-way 5-shot

```python
sizes = np.array([0, 5, 10, 15, 20, 25, 30])

padding_mean = np.array([63.04, 73.31, 73.05, 72.72, 73.05, 72.05, 73.05])
fixed_mean   = np.array([63.04, 72.55, 72.95, 73.10, 72.05, 73.05, 73.05])
random_mean  = np.array([63.04, 72.36, 72.45, 73.07, 73.05, 73.05, 72.05])

#  
plt.figure(figsize=(6, 4))
plt.plot(sizes, padding_mean, '-o', label='Padding',linewidth=2.5)
plt.plot(sizes, fixed_mean, '-s', label='Fixed Patch', linewidth=2.5)
plt.plot(sizes, random_mean, '-^', label='Random Patch', linewidth=2.5)

# , , 
# plt.title('Impact of Prompt Size and Template on miniImageNet (5-way 5-shot)', fontsize=12)
# plt.ylim(60.0, 80.0)
plt.xlabel('Prompt Size', fontsize=16)
plt.ylabel('Accuracy (%)', fontsize=16)
plt.xticks(sizes, fontsize=15)
plt.grid(alpha=0.5, linestyle='--')
plt.legend(fontsize=15)

#    
plt.tight_layout()
plt.savefig("5way_5shot_promptsize.png", dpi=300, bbox_inches='tight')
plt.show()
```
