```python
import pandas as pd
import matplotlib.pyplot as plt


reguarization_csv = '../MAML_Prompt_padding_5way_5shot_filter128_miniImagenet/logs/summary_statistics.csv'
not_reguarization_csv = '../MAML_Prompt_padding_5way_5shot_filter128_miniImagenet(DataAugNone)/logs/summary_statistics.csv'

# CSV 불러오기
df_reg = pd.read_csv(reguarization_csv)
df_no_reg = pd.read_csv(not_reguarization_csv)
```

```python
# 그래프 그리기
plt.figure(figsize=(8, 5))


# 정규화 적용
plt.plot(df_reg['epoch'], df_reg['train_loss_mean'], label='Train Loss (Reg)', color='orange', linewidth=2)
plt.plot(df_reg['epoch'], df_reg['val_loss_mean'], label='Val Loss (Reg)', color='orange', linestyle='--', linewidth=2)

# 정규화 미적용
plt.plot(df_no_reg['epoch'], df_no_reg['train_loss_mean'], label='Train Loss (No Reg)', color='SkyBlue', linewidth=2)
plt.plot(df_no_reg['epoch'], df_no_reg['val_loss_mean'], label='Val Loss (No Reg)', color='SkyBlue', linestyle='--', linewidth=2)


# 라벨 및 제목 폰트 크기 설정
plt.xlabel('Epoch', fontsize=16)
plt.ylabel('Loss', fontsize=16)
plt.ylabel('Loss')
plt.title('Train vs Validation Loss Mean Over Epochs')
plt.legend(fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 그래프 그리기
plt.figure(figsize=(8, 5))

# 정규화 적용 (주황색, 실선)
plt.plot(df_reg['epoch'], df_reg['train_accuracy_mean'], label='Train Acc (Reg)', color='orange', linewidth=2)
plt.plot(df_reg['epoch'], df_reg['val_accuracy_mean'], label='Val Acc (Reg)', color='orange', linewidth=2)

# 정규화 미적용 (연한 파란색, 실선)
plt.plot(df_no_reg['epoch'], df_no_reg['train_accuracy_mean'], label='Train Acc (No Reg)', color='lightblue', linewidth=2)
plt.plot(df_no_reg['epoch'], df_no_reg['val_accuracy_mean'], label='Val Acc (No Reg)', color='lightblue', linewidth=2)

# 시각화 설정
plt.xlabel('Epoch', fontsize=16)
plt.ylabel('Accuracy', fontsize=16)
plt.title('Train and Validation Accuracy Comparison (With vs Without Regularization)', fontsize=17)
plt.legend(fontsize=13)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.grid(True)
plt.tight_layout()
plt.show()
```
