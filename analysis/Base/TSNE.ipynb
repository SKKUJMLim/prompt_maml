```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# 랜덤 데이터 생성 (300개 샘플, 100차원)
np.random.seed(42)
n_samples = 300
n_features = 100

data = np.random.randn(n_samples, n_features)  # 100차원의 랜덤 데이터
labels = np.arange(n_samples)  # 각 데이터 포인트에 고유한 레이블 (0 ~ 299)

# t-SNE 모델
tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000)
data_tsne = tsne.fit_transform(data)

# 시각화
plt.figure(figsize=(10, 8))

# colormap 선택: 'rainbow', 'hsv', 'nipy_spectral', 'tab20'
scatter = plt.scatter(data_tsne[:, 0], data_tsne[:, 1], c=labels, cmap='nipy_spectral', s=20)
plt.colorbar(scatter, label='Labels')  # 컬러바 추가
plt.title("t-SNE Visualization with Colorful Labels")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.show()
```

```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# 랜덤 데이터 생성 (3개의 클러스터)
np.random.seed(42)  # 재현성을 위해 랜덤 시드 설정
n_samples = 300
n_features = 50

# 클러스터 중심 생성
cluster_centers = [np.random.randn(n_features) for _ in range(3)]

# 각 클러스터에 데이터 추가
data = np.vstack([
    cluster_centers[i] + np.random.randn(n_samples // 3, n_features) * 0.5
    for i in range(3)
])
labels = np.array([i for i in range(3) for _ in range(n_samples // 3)])

# t-SNE 모델
tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000)
data_tsne = tsne.fit_transform(data)

# 시각화
plt.figure(figsize=(8, 6))
scatter = plt.scatter(data_tsne[:, 0], data_tsne[:, 1], c=labels, cmap='tab10', s=10)
plt.colorbar(scatter, ticks=[0, 1, 2], label='Clusters')
plt.title("t-SNE Visualization with Random Data")
plt.show()
```
