# 1. Gaussian_noise

```python
import matplotlib.pyplot as plt
import numpy as np

# x축 (noise std)
std_values = [0.0, 0.02, 0.05, 0.08, 0.10]

# 각 방법의 평균 ± 표준편차 데이터
methods = {
    "MAML": {"mean": [], "std": []},
    "ANIL": {"mean": [67.03, 66.23, 62.41, 59.86, 56.68],
             "std":  [0.47, 0.47, 0.48, 0.49, 0.49]},
    "BOIL": {"mean": [], "std": []},
    "L2F":  {"mean": [], "std": []},
    "ALFA": {"mean": [], "std": []},
    "GAP":  {"mean": [71.51, 69.83, 65.96, 59.54, 54.48],
             "std":  [0.61, 0.46, 0.47, 0.49, 0.50]},
    "DCML (Ours)": {"mean": [73.31, 72.09, 68.01, 64.23, 61.88],
                    "std":  [0.44, 0.45, 0.47, 0.48, 0.48]},
}



# 그래프 그리기
plt.figure(figsize=(9, 5))

for name, data in methods.items():
    if len(data["mean"]) > 0:
        plt.errorbar(std_values, data["mean"], yerr=data["std"],
                     marker='o', capsize=4, linewidth=2, label=name)
    else:
        # 값이 없는 방법은 legend만 표시
        plt.plot([], [], label=name, linestyle='--', color='gray', alpha=0.5)

# 축 및 제목
plt.xlabel("Noise Standard Deviation (std)", fontsize=12)
plt.ylabel("Accuracy (%)", fontsize=12)
plt.title("Performance under Different Noise Levels", fontsize=14)
plt.xticks(std_values)
plt.yticks(np.arange(40, 71, 10))  # 20~70, 10 단위
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

# 2. Impulse Input Noise

```python
import matplotlib.pyplot as plt
import numpy as np

# x축 (noise std)
std_values = [0.0, 0.1, 0.2, 0.3, 0.4]

# 각 방법의 평균 ± 표준편차 데이터
methods = {
    "MAML": {"mean": [], "std": []},
    "ANIL": {"mean": [66.03, 47.42, 36.91, 31.26, 28.63],
             "std":  [0.47, 0.50, 0.48, 0.46, 0.45]},
    "BOIL": {"mean": [], "std": []},
    "L2F":  {"mean": [], "std": []},
#     "ALFA": {"mean": [69.12, 63.88, 44.22, 38.00, 33.12],
#              "std":  [0.47, 0.48, 0.50, 0.49, 0.47]},
    "GAP":  {"mean": [71.51, 61.86, 34.75, 30.78, 27.87],
             "std":  [0.61, 0.49, 0.48, 0.46, 0.45]},
    "DCML (Ours)": {"mean": [73.31, 63.58, 47.29, 41.71, 37.42],
                    "std":  [0.44, 0.48, 0.50, 0.49, 0.48]},
}

# 그래프 그리기
plt.figure(figsize=(8, 5))

for name, data in methods.items():
    if len(data["mean"]) > 0:  # 값이 있는 경우에만 선 그리기
        plt.errorbar(std_values, data["mean"], yerr=data["std"],
                     marker='o', capsize=4, linewidth=2.5, label=name)
    else:
        # 값이 없는 항목은 legend에만 표시
        plt.plot([], [], label=name, linestyle='--', color='gray', alpha=0.5)

# 축 및 제목 설정
save_path = "Impulse Noise Strength"
plt.xlabel("Impulse Noise Strength", fontsize=16)
plt.ylabel("Accuracy (%)", fontsize=15)
plt.title("Impulse Input Noise", fontsize=15)
plt.xticks(std_values)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=10)
plt.tight_layout()
plt.savefig(save_path, dpi=300, bbox_inches='tight')
plt.show()
```

# 3. Uniform Input Noise

```python
import matplotlib.pyplot as plt
import numpy as np

# x축 (noise std)
std_values = [0.0, 0.1, 0.2, 0.3, 0.4]

# 각 방법의 평균 ± 표준편차 데이터
methods = {
    "MAML": {
        "mean": [],  # 아직 값 없음
        "std": []
    },
    "L2F": {
        "mean": [],  # 아직 값 없음
        "std": []
    },
#     "ALFA": {
#         "mean": [69.12, 60.14, 49.80, 41.61, 36.40],
#         "std": [0.47, 0.49, 0.50, 0.49, 0.48]
#     },
    "GAP": {
        "mean": [71.51, 54.41, 38.96, 32.72, 29.09],
        "std": [0.61, 0.50, 0.49, 0.46, 0.45]
    },
    "DCML (Ours)": {
        "mean": [73.31, 61.38, 51.72, 43.97, 38.74],
        "std": [0.44, 0.49, 0.50, 0.50, 0.00]
    }
}

# 그래프 그리기
plt.figure(figsize=(8, 5))

for name, data in methods.items():
    if len(data["mean"]) > 0:  # 값이 있는 경우에만 그리기
        plt.errorbar(std_values, data["mean"][:len(std_values)], 
                     yerr=data["std"][:len(std_values)], 
                     marker='o', capsize=4, linewidth=2, label=name)
    else:
        # 빈 데이터라도 범례에는 추가
        plt.plot([], [], label=name, linestyle='--', color='gray', alpha=0.5)

# 축 및 제목 설정
save_path = "Uniform Noise Strength"
plt.xlabel("Uniform Noise Strength", fontsize=16)
plt.ylabel("Accuracy (%)", fontsize=15)
plt.title("Impulse Input Noise", fontsize=15)
plt.xticks(std_values)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=10)
plt.tight_layout()
plt.savefig(save_path, dpi=300, bbox_inches='tight')
plt.show()
```
