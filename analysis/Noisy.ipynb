# 1. Gaussian_noise

```python
import matplotlib.pyplot as plt
import numpy as np

# x (noise std)
std_values = [0.0, 0.05, 0.10, 0.15, 0.20]

#      
methods = {
    "MAML": {"mean": [], "std": []},
    "ANIL": {"mean": [66.03, 62.41, 56.68, 51.03, 46.32],
             "std":  [0.47, 0.48, 0.49, 0.50, 0.50]},
    "BOIL": {"mean": [], "std": []},
    "L2F":  {"mean": [], "std": []},
#     "ALFA": {"mean": [69.12, 65.96, 54.48, 45.01, 39.80],
#              "std":  [0.47, 0.47, 0.50, 0.50, 0.48]},
    "GAP":  {"mean": [71.51, 65.96, 54.48, 45.01, 39.80],
             "std":  [0.61, 0.47, 0.50, 0.50, 0.48]},
    "DCML (Ours)": {"mean": [73.31, 68.01, 61.88, 56.79, 52.56],
                    "std":  [0.44, 0.47, 0.48, 0.50, 0.50]},
}

#  
plt.figure(figsize=(9, 5))

for name, data in methods.items():
    if len(data["mean"]) > 0:
        plt.errorbar(std_values, data["mean"], yerr=data["std"],
                     marker='o', capsize=4, linewidth=2, label=name)
    else:
        #    legend 
        plt.plot([], [], label=name, linestyle='--', color='gray', alpha=0.5)

#   
plt.xlabel("Noise Standard Deviation (std)", fontsize=12)
plt.ylabel("Accuracy (%)", fontsize=12)
plt.title("Performance under Different Noise Levels", fontsize=14)
plt.xticks(std_values)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

# 2. Impulse Input Noise

```python
import matplotlib.pyplot as plt
import numpy as np

# x (noise std)
std_values = [0.0, 0.1, 0.2, 0.3, 0.4]

#      
methods = {
    "MAML": {"mean": [], "std": []},
    "ANIL": {"mean": [66.03, 47.42, 36.91, 31.26, 28.63],
             "std":  [0.47, 0.50, 0.48, 0.46, 0.45]},
    "BOIL": {"mean": [], "std": []},
    "L2F":  {"mean": [], "std": []},
#     "ALFA": {"mean": [69.12, 63.88, 44.22, 38.00, 33.12],
#              "std":  [0.47, 0.48, 0.50, 0.49, 0.47]},
    "GAP":  {"mean": [71.51, 61.86, 34.75, 30.78, 27.87],
             "std":  [0.61, 0.49, 0.48, 0.46, 0.45]},
    "DCML (Ours)": {"mean": [73.31, 63.58, 47.29, 41.71, 37.42],
                    "std":  [0.44, 0.48, 0.50, 0.49, 0.48]},
}

#  
plt.figure(figsize=(8, 5))

for name, data in methods.items():
    if len(data["mean"]) > 0:  #     
        plt.errorbar(std_values, data["mean"], yerr=data["std"],
                     marker='o', capsize=4, linewidth=2.5, label=name)
    else:
        #    legend 
        plt.plot([], [], label=name, linestyle='--', color='gray', alpha=0.5)

#    
save_path = "Impulse Noise Strength"
plt.xlabel("Impulse Noise Strength", fontsize=16)
plt.ylabel("Accuracy (%)", fontsize=15)
plt.title("Impulse Input Noise", fontsize=15)
plt.xticks(std_values)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=10)
plt.tight_layout()
plt.savefig(save_path, dpi=300, bbox_inches='tight')
plt.show()
```
