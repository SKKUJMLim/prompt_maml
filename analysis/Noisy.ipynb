# 1. Gaussian_noise

```python
import matplotlib.pyplot as plt
import numpy as np

# x (noise std)
std_values = [0.0, 0.02, 0.05, 0.08, 0.10]

#      
methods = {
    "MAML": {
        "mean": [65.33, 64.78, 61.74, 56.37, 52.46],
        "std":  [0.48, 0.48, 0.49, 0.50, 0.50]
    },
    "ANIL": {
        "mean": [64.03, 63.23, 59.41, 56.86, 53.68],
        "std":  [0.47, 0.47, 0.48, 0.49, 0.49]
    },
    "GAP": {
        "mean": [71.51, 69.83, 65.96, 59.54, 54.48],
        "std":  [0.61, 0.46, 0.47, 0.49, 0.50]
    },
    "DCML (Ours)": {
        "mean": [73.31, 72.09, 68.01, 64.23, 61.88],
        "std":  [0.44, 0.45, 0.47, 0.48, 0.48]
    }
}

#   (  )
colors = {
    "MAML": "#1f77b4",
    "ANIL": "#2ca02c",
    "GAP": "#17becf",
    "DCML (Ours)": "#fd8d3c"
}

#  
plt.figure(figsize=(8, 5))

for name, data in methods.items():
    linestyle = '-' if name == "DCML (Ours)" else '--'  # DCML 
    plt.errorbar(std_values, data["mean"], yerr=data["std"],
                 marker='o', capsize=3, linewidth=2,
                 linestyle=linestyle, label=name,
                 color=colors.get(name, None))

#    
plt.xlabel("Gaussian Noise Strength")
plt.ylabel("Accuracy (%)")
plt.title("Gaussian Input Noise")
plt.xticks(std_values)
plt.yticks(np.arange(50, 76, 5))
plt.ylim(50, 75)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(loc="upper right")

#   
plt.tight_layout()
plt.savefig("Gaussian_Noise_Strength.png", dpi=300, bbox_inches='tight')
plt.show()
```

# 2. Impulse Input Noise

```python
import matplotlib.pyplot as plt
import numpy as np

# x (noise std)
std_values = [0.0, 0.1, 0.2, 0.3, 0.4]

#      
methods = {
    "MAML": {"mean": [], "std": []},
    "ANIL": {"mean": [66.03, 47.42, 36.91, 31.26, 28.63],
             "std":  [0.47, 0.50, 0.48, 0.46, 0.45]},
    "BOIL": {"mean": [], "std": []},
    "L2F":  {"mean": [], "std": []},
#     "ALFA": {"mean": [69.12, 63.88, 44.22, 38.00, 33.12],
#              "std":  [0.47, 0.48, 0.50, 0.49, 0.47]},
    "GAP":  {"mean": [71.51, 61.86, 34.75, 30.78, 27.87],
             "std":  [0.61, 0.49, 0.48, 0.46, 0.45]},
    "DCML (Ours)": {"mean": [73.31, 63.58, 47.29, 41.71, 37.42],
                    "std":  [0.44, 0.48, 0.50, 0.49, 0.48]},
}

#  
plt.figure(figsize=(8, 5))

for name, data in methods.items():
    if len(data["mean"]) > 0:  #     
        plt.errorbar(std_values, data["mean"], yerr=data["std"],
                     marker='o', capsize=4, linewidth=2.5, label=name)
    else:
        #    legend 
        plt.plot([], [], label=name, linestyle='--', color='gray', alpha=0.5)

#    
save_path = "Impulse Noise Strength"
plt.xlabel("Impulse Noise Strength", fontsize=16)
plt.ylabel("Accuracy (%)", fontsize=15)
plt.title("Impulse Input Noise", fontsize=15)
plt.xticks(std_values)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=10)
plt.tight_layout()
plt.savefig(save_path, dpi=300, bbox_inches='tight')
plt.show()
```
