# 1. Gaussian_noise

```python
import matplotlib.pyplot as plt
import numpy as np

# x축 (noise std)
std_values = [0.0, 0.02, 0.04, 0.06, 0.08, 0.10]

# 각 방법의 평균 ± 표준편차 데이터
methods = {
    "MAML": {
        "mean": [65.33, 64.78, 63.02, 60.12, 56.37, 52.46],
        "std":  [0.48, 0.48, 0.48, 0.49, 0.50, 0.50]
    },
    "ANIL": {
        "mean": [64.03, 63.23, 60.54, 59.65, 56.86, 53.68],
        "std":  [0.47, 0.47, 0.48, 0.49, 0.49, 0.49]
    },
    "GAP": {
        "mean": [71.51, 69.83, 67.42, 64.08, 59.54, 54.48],
        "std":  [0.61, 0.46, 0.47, 0.48, 0.49, 0.50]
    },
    "DCML (Our)": {
        "mean": [73.31, 72.09, 69.43, 66.64, 64.23, 61.88],
        "std":  [0.44, 0.45, 0.46, 0.47, 0.48, 0.48]
    }
}

# 🎨 색상 팔레트 (DCML = 주황, 나머지 점선)
colors = {
    "MAML": "#1f77b4",       # 파랑
    "ANIL": "#2ca02c",       # 녹청
    "GAP": "#17becf",        # 청록
    "DCML (Our)": "#ff7f0e"  # 🟠 주황 (제안 방법)
}

# 그래프 그리기
plt.figure(figsize=(8, 5))

for name, data in methods.items():
    linestyle = '-' if name == "DCML (Our)" else '--'  # DCML만 실선
    plt.errorbar(std_values, data["mean"], yerr=data["std"],
                 marker='o', capsize=3, linewidth=2,
                 linestyle=linestyle, label=name,
                 color=colors.get(name, None))

# 축 및 제목 설정
plt.xlabel("Gaussian Noise Strength")
plt.ylabel("Accuracy (%)")
plt.title("Performance under Gaussian Input Noise")
plt.xticks(std_values)
plt.yticks(np.arange(50, 76, 5))
plt.ylim(50, 75)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(loc="upper right")

# 저장 및 표시
plt.tight_layout()
plt.savefig("Gaussian_Noise_Strength.png", dpi=300, bbox_inches='tight')
plt.show()
```
