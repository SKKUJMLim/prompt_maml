```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 파일 경로 설정
file1_path = "../MAML_Prompt_padding_5way_5shot_filter128_miniImagenet_per_task_acc.csv"
file2_path = "../MAML_5way_5shot_filter128_miniImagenet_per_task_acc.csv"

# CSV 파일 불러오기 (header 없음 처리)
df1 = pd.read_csv(file1_path, header=None)
df2 = pd.read_csv(file2_path, header=None)

# 첫 번째 행을 column 이름으로 설정
df1.columns = df1.iloc[0]
df2.columns = df2.iloc[0]
df1 = df1.drop(index=0).reset_index(drop=True)
df2 = df2.drop(index=0).reset_index(drop=True)

# 분석할 column 지정
target_column = 'layer_dict.conv0.conv.weight_pairwise_cos_sim'

# 숫자형으로 변환 + 결측값 제거
cos_sim_1 = pd.to_numeric(df1[target_column], errors='coerce').dropna().reset_index(drop=True)
cos_sim_2 = pd.to_numeric(df2[target_column], errors='coerce').dropna().reset_index(drop=True)

# 그래프 그리기
plt.figure(figsize=(10, 5))
plt.plot(cos_sim_1, label='MAML + Prompt', marker='o')
plt.plot(cos_sim_2, label='MAML', marker='s')
plt.xlabel('Iteration')
plt.ylabel('Cosine Similarity (conv0)')
plt.title('Pairwise Cosine Similarity of conv0 across Iterations')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 파일 경로
file1_path = "../MAML_Prompt_padding_5way_5shot_filter128_miniImagenet_per_task_acc.csv"
file2_path = "../MAML_5way_5shot_filter128_miniImagenet_per_task_acc.csv"

# CSV 불러오기
df1 = pd.read_csv(file1_path, header=None)
df2 = pd.read_csv(file2_path, header=None)

# 첫 행을 컬럼명으로 설정
df1.columns = df1.iloc[0]
df2.columns = df2.iloc[0]
df1 = df1.drop(index=0).reset_index(drop=True)
df2 = df2.drop(index=0).reset_index(drop=True)

# 분석 대상 컬럼
target_column = 'layer_dict.conv0.conv.weight_cos_sim'

# 숫자형 변환 및 결측치 제거
cos_sim_1 = pd.to_numeric(df1[target_column], errors='coerce').dropna().reset_index(drop=True)
cos_sim_2 = pd.to_numeric(df2[target_column], errors='coerce').dropna().reset_index(drop=True)

# 그래프 시각화 (두 개의 subplot 분리)
fig, axs = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# MAML + Prompt
axs[0].plot(cos_sim_1, label='MAML + Prompt', marker='o', color='tab:blue')
axs[0].set_ylim(0, 1)
axs[0].set_ylabel('Cosine Similarity (conv0)')
axs[0].set_title('MAML + Prompt')
axs[0].grid(True)
axs[0].legend()

# MAML
axs[1].plot(cos_sim_2, label='MAML', marker='s', color='tab:orange')
axs[1].set_ylim(0, 1)
axs[1].set_xlabel('Iteration')
axs[1].set_ylabel('Cosine Similarity (conv0)')
axs[1].set_title('MAML')
axs[1].grid(True)
axs[1].legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 파일 경로
file1_path = "../MAML_Prompt_padding_5way_5shot_filter128_miniImagenet_per_task_acc.csv"
file2_path = "../MAML_5way_5shot_filter128_miniImagenet_per_task_acc.csv"

# CSV 불러오기
df1 = pd.read_csv(file1_path, header=None)
df2 = pd.read_csv(file2_path, header=None)

# 첫 행을 컬럼명으로 설정
df1.columns = df1.iloc[0]
df2.columns = df2.iloc[0]
df1 = df1.drop(index=0).reset_index(drop=True)
df2 = df2.drop(index=0).reset_index(drop=True)

# 분석 대상 컬럼
target_column = 'layer_dict.conv0.conv.weight_dot_product'

# 숫자형 변환 및 결측치 제거
dot1 = pd.to_numeric(df1[target_column], errors='coerce').dropna().reset_index(drop=True)
dot2 = pd.to_numeric(df2[target_column], errors='coerce').dropna().reset_index(drop=True)

# 이상치 제거 (상위 1% 이상 제거)
threshold_1 = dot1.quantile(0.99)
threshold_2 = dot2.quantile(0.99)
dot1_clipped = dot1[dot1 <= threshold_1].reset_index(drop=True)
dot2_clipped = dot2[dot2 <= threshold_2].reset_index(drop=True)

# Y축 범위 맞추기
y_min = min(dot1_clipped.min(), dot2_clipped.min())
y_max = max(dot1_clipped.max(), dot2_clipped.max())

# 그래프 시각화 (분리된 subplot + 동일 y축 범위)
fig, axs = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

axs[0].plot(dot1_clipped, label='MAML + Prompt', marker='o', color='tab:blue')
axs[0].set_ylim(y_min, y_max)
axs[0].set_ylabel('Dot Product (conv0)')
axs[0].set_title('MAML + Prompt (Outliers Removed)')
axs[0].grid(True)
axs[0].legend()

axs[1].plot(dot2_clipped, label='MAML', marker='s', color='tab:orange')
axs[1].set_ylim(y_min, y_max)
axs[1].set_xlabel('Iteration')
axs[1].set_ylabel('Dot Product (conv0)')
axs[1].set_title('MAML (Outliers Removed)')
axs[1].grid(True)
axs[1].legend()

plt.tight_layout()
plt.show()
```
