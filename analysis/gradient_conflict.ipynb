```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# CSV 파일 불러오기
file_path = '../MAML_Prompt_padding_5way_5shot_filter128_miniImagenet/MAML_Prompt_padding_5way_5shot_filter128_miniImagenet_per_task_acc.csv'
df = pd.read_csv(file_path)

# cos_sim 시각화
plt.figure(figsize=(10, 6))
plt.plot(df['phase'], df['cos_sim'], marker='o', linestyle='-')
plt.title('Cosine Similarity over Training Phases')
plt.xlabel('Phase')
plt.ylabel('Cosine Similarity')
plt.grid(True)
plt.tight_layout()
plt.show()
```
