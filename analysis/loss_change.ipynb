```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# CSV 파일 경로 설정
file_path_with_regularization = "../MAML_Prompt_padding_5way_5shot_filter128_miniImagenet/logs/summary_statistics.csv"
file_path_without_regularization = "../MAML_Prompt_padding_5way_5shot_filter64_miniImagenet/logs/summary_statistics.csv"
file_path_MAML = "../MAML_5way_5shot_filter64_miniImagenet/logs/summary_statistics.csv"
```

```python
# 파일 읽기
def load_and_filter_csv(path):
    df = pd.read_csv(path)
    return df.dropna(subset=["epoch", "train_loss_mean", "val_loss_mean",
                             "train_accuracy_mean", "val_accuracy_mean"])

df_with_reg = load_and_filter_csv(file_path_with_regularization)
df_without_reg = load_and_filter_csv(file_path_without_regularization)
df_maml = load_and_filter_csv(file_path_MAML)

# 정확도 단위를 %로 변환
for df in [df_with_reg, df_without_reg, df_maml]:
    df["train_accuracy_mean"] *= 100
    df["val_accuracy_mean"] *= 100
```

## Train loss와 Validation loss 비교

```python
plt.figure(figsize=(10, 6))
plt.plot(df_with_reg["epoch"], df_with_reg["train_loss_mean"],
         label="Train Loss (With Reg)", color='red', linestyle='-')
plt.plot(df_with_reg["epoch"], df_with_reg["val_loss_mean"],
         label="Val Loss (With Reg)", color='red', linestyle='--')

plt.plot(df_without_reg["epoch"], df_without_reg["train_loss_mean"],
         label="Train Loss (No Reg)", color='blue', linestyle='-')
plt.plot(df_without_reg["epoch"], df_without_reg["val_loss_mean"],
         label="Val Loss (No Reg)", color='blue', linestyle='--')

# plt.plot(df_maml["epoch"], df_maml["train_loss_mean"],
#          label="Train Loss (MAML)", color='green', linestyle='-')
# plt.plot(df_maml["epoch"], df_maml["val_loss_mean"],
#          label="Val Loss (MAML)", color='green', linestyle='--')

plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Train/Validation Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Train Accuracy와 Validation Accuracy 비교

```python
plt.figure(figsize=(10, 6))
plt.plot(df_with_reg["epoch"], df_with_reg["train_accuracy_mean"],
         label="Train Acc (With Reg)", color='red', linestyle='-')
plt.plot(df_with_reg["epoch"], df_with_reg["val_accuracy_mean"],
         label="Val Acc (With Reg)", color='red', linestyle='--')

plt.plot(df_without_reg["epoch"], df_without_reg["train_accuracy_mean"],
         label="Train Acc (No Reg)", color='blue', linestyle='-')
plt.plot(df_without_reg["epoch"], df_without_reg["val_accuracy_mean"],
         label="Val Acc (No Reg)", color='blue', linestyle='--')

# plt.plot(df_maml["epoch"], df_maml["train_accuracy_mean"],
#          label="Train Acc (MAML)", color='green', linestyle='-')
# plt.plot(df_maml["epoch"], df_maml["val_accuracy_mean"],
#          label="Val Acc (MAML)", color='green', linestyle='--')

plt.xlabel("Epoch")
plt.ylabel("Accuracy (%)")
plt.title("Train/Validation Accuracy")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
