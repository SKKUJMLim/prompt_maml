```python
import pandas as pd
import matplotlib.pyplot as plt
import torch
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
```

```python
# Load the uploaded CSV file to examine its structure
file_path = '../MAML_Prompt_fixed_patch_5way_5shot_filter64_miniImagenet/MAML_Prompt_fixed_patch_5way_5shot_filter64_miniImagenet_per_task_acc.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the data to understand its structure
data.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data.index, data["accuracy"], marker='o', linestyle='-', label="Accuracy")
plt.title("Accuracy Trend Across All Tasks (fixed_patch)", fontsize=14)
plt.xlabel("Index of tasks", fontsize=12)
plt.ylabel("Accuracy", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```
